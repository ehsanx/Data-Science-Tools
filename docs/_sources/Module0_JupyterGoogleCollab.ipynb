1. **Code**: Allows for the execution of code in multiple languages.

## Kernels & Code Cells

Code cells contain executable code written in a specific programming language.

Each notebook is connected to a kernel, the backend engine responsible for running the code.

For instance, a Jupyter notebook can be connected to a Python kernel. One can install a kernel such as the ipykernel, which is the interactive Python engine that works with JupyterLab. When you create a notebook with the kernel from a particular Conda environment, you'll have access to all the packages installed in that environment, similar to running Python code directly in that environment.

Code cells are executed using the notebook's kernel. When you run a code cell, the code is sent to the kernel, which executes it. Any outputs are sent back to the notebook and displayed as the cell's output. You can run code cells in any order.

Each time you run a code cell, the number to the left of the cell (e.g., [8]:) increments by 1, helping you track the execution order.


## Running code cells

1. Ensure that the cell is selected (click anywhere within the code cell to select it)
2. To run the cell:
- Click the little play button (on the left side of the code block in Collab, or in the top toolbar in Jupyter)
- Alternatively, one can hit `Shift + Enter` to run the cell, then automatically move on to the next one below.

## Markdown Cells

So far, we've discussed code cells, which handle the computational aspects of a ipynb notebook. The other key type of cell commonly found within these notebooks are the Markdown cell, which allows you to write formatted text. This is where you can include the text-heavy and narrative portion of your notebook.

Markdown offers a quick and easy way to format plain text without needing a graphical interface/special file format, and can be used in standalone text files (typically with a .md extension, like README.md) or within frameworks such as Jupyter & Collab notebooks.

Markdown offers several key features: it allows for the quick writing of text that is nicely formatted upon rendering. Moreover, its content remains simply as plain text, which can be stored much easier.

## Markdown Basics

You can make text *italic* or **bold** by surrounding a block of text with a single or double * respectively.

You can also build nested itemized or enumerated lists:

* One
    - Sublist
        - This
  - Sublist
        - That
        - The other thing
* Two
  - Sublist
* Three
  - Sublist



You can create headings by starting a line with one or more # symbols followed by a space. Heres an example:

```
# Heading 1
## Heading 2
### Heading 2.1
### Heading 2.2
```
