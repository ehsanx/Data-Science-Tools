## What are Github Branches?

GitHub branches allow developers to create separate versions of a repository to work on different features or fixes concurrently. The default branch is usually called "main" or "master," and new branches can be created from it to isolate changes.

This enables multiple collaborators to work on their tasks without affecting the main codebase until the changes are ready to be merged. Once the work on a branch is complete, it can be reviewed and merged back into the main branch, integrating the new features or fixes.

To check what branch you're currently on, execute `git branch` in the command line.

This command will list all the branches in the repository and highlight the current branch with an asterisk (*) next to it.

For instance (below branches are just examples):

```
$ git branch
* main
  feature-branch
  bugfix-branch
```

Alternatively, you can use `git status` to display the current branch along with some other useful information:

```
On branch main
Your branch is up to date with 'origin/main'.

nothing to commit, working tree clean
```

## Going back to new Branch


Let's go back to the `new` branch, in which you will see a message saying that we need to stash our changes.

image.png

In Git, stashes allow developers to temporarily save changes in a working directory that are not ready to be committed. This is useful when you need to switch branches or perform other tasks without losing your current work.

When you stash changes, Git saves the modified tracked files and staged changes in a stack-like storage, allowing you to reapply them later. You can then retrieve and reapply these stashed changes, enabling seamless transitions between different tasks without committing unfinished work.

To view the list of stashed changes in your Git repository, you can use the following command:

```
git stash list
```

After clicking "View Stash", we get brought to the following page.
At this step, let us click "Restore".

image.png

To restore a stash in Git, you use the `git stash apply` or `git stash pop` command. The `apply` command re-applies the stashed changes to your working directory but keeps the stash in the stash list, while the `pop` command re-applies the changes and removes the stash from the list.

After clicking Restore, let's go ahead and check what our VSCode looks like.

Keep in mind that we are on the `new` branch, as we open up our VSCode.

image.png

We can see that now, the newly added `NewBranch.txt` shows up in our directory.

This is because we restored our stashed changes.

To push the branch onto Github, use the `git push` command to push the branch to the remote repository:

```
git push origin branch-name
```

If this is the first time you are pushing this branch and you want to set the upstream branch, you can use the `--set-upstream` (or `-u`) flag:


```
git push -u origin branch-name
```
